```python
from torch_geometric.utils import is_undirected
from torch_geometric.data import Data, DataLoader
import torch
import numpy as np
from torch_scatter import scatter_add
```

```python
# create data with multi-dimensionality random walk featues 

drug_data_list = list([] for i in range(drug_gene_list[0].shape[0]))
food_data_list = list([] for i in range(food_gene_list[0].shape[0]))

for i in range(len(drug_gene_list)):
    for idx, vector in enumerate(drug_gene_list[i]):
        drug_data_list[idx].append(torch.from_numpy(vector))                

for i in range(len(food_gene_list)):
    for idx, vector in enumerate(food_gene_list[i]):
        food_data_list[idx].append(torch.from_numpy(vector))

new_drug_data_list = []
for data in drug_data_list:
    new_drug_data_list.append(torch.stack(data, dim=0).t())
new_food_data_list = []
for data in food_data_list:
    new_food_data_list.append(torch.stack(data, dim=0).t())
```

```python
# generate pathway_edge_index from pathway matrix
import torch
pathway_mat = torch.from_numpy(pathway.values)
pathway_edge_index = pathway_mat.nonzero().t()

pathway_kegg_mat = torch.from_numpy(pathway_kegg.values)
pathway_kegg_edge_index = pathway_kegg_mat.nonzero().t()
```

```python
pathway_edge_index.shape
```

```python
# create torch_geometric.data.Data list
def create_data_list(node_feature_mat, disease=None):
    # list: node_feature_mat with dim [num_drugs, -]
    # torch.Tensor for each feature with dim [num_nodes, feature_dim]
    
    data_list = []
    if disease is not None:
        for feature, dis in zip(node_feature_mat, disease):
            data_list.append(Data(x=feature.type(torch.float32), edge_index=edge_index, edge_attr=edge_attr.type(torch.float32), y=torch.from_numpy(dis).type(torch.float32)))
        return data_list
    else:
        for feature in node_feature_mat:
            data_list.append(Data(x=feature.type(torch.float32), edge_index=edge_index, edge_attr=edge_attr.type(torch.float32)))
        return data_list

drug_gene_data = create_data_list(new_drug_data_list, disease.values)
food_gene_data = create_data_list(new_food_data_list)
```

```python
### split train/test set for binary classification###
anticancer_disease = disease.values[:, 12]

labeled_dis = torch.from_numpy(anticancer_disease.nonzero()[0])
non_labeled_dis = torch.from_numpy((anticancer_disease == 0).nonzero()[0])

def split_dataset(indices, ratio=0.8):
    index = torch.randperm(len(indices))
    num_train = int(len(indices) * 0.8)
    train_index = index[:num_train]
    test_index = index[num_train:]
    print('n_train: {}'.format(len(train_index)))
    print('n_test: {}'.format(len(test_index)))
    return indices[train_index], indices[test_index]

train_labeled_drug, test_labeld_drug = split_dataset(labeled_dis, 0.8)
train_non_labeled_drug, test_non_labeld_drug = split_dataset(non_labeled_dis, 0.8)

train_drug = torch.cat((train_labeled_drug, train_non_labeled_drug), dim=0).numpy()
test_drug = torch.cat((test_labeld_drug, test_non_labeld_drug), dim=0).numpy()

train_mask_fp = './dataset/processed/train_mask_binary.npy'
test_mask_fp = './dataset/processed/test_mask_binary.npy'

np.save(train_mask_fp, train_drug[torch.randperm(len(train_drug))])
np.save(test_mask_fp, test_drug[torch.randperm(len(test_drug))])
```

#### Computing node embeddings of ppi to use as positional features in GCNGraphLets and MoNet

```python
import torch
from torch_geometric.nn import Node2Vec
from torch.utils.data import DataLoader
```

```python
edge_index = torch.load('./dataset/processed/new/raw_ac.py')[0].edge_index
```

```python
##Code adapted from https://github.com/rusty1s/pytorch_geometric/blob/master/examples/node2vec.py 
# to obtain node embeddings for PPI
n_genes = 15135


loader = DataLoader(torch.arange(n_genes), batch_size=128, shuffle=True)

device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = Node2Vec(n_genes, embedding_dim=10, walk_length=4,
                 context_size=4, walks_per_node=10)
edge_index = edge_index.to(device)
model= model.to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=0.1)


def train():
    model.train()
    total_loss = 0
    for subset in loader:
        optimizer.zero_grad()
        loss = model.loss(edge_index, subset.to(device))
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(loader)


scheduler = torch.optim.lr_scheduler.StepLR(optimizer, 1, gamma=0.98)

```

```python
for epoch in range(1, 100):
    loss = train()
    scheduler.step()
    print('Epoch: {:02d}, Loss: {:.4f}'.format(epoch, loss))
```

```python
model.eval()
graph_emb = model(torch.arange(n_genes, device=device))
```

```python
# create data with single feature 

ac_data_list = list([] for i in range(ac_gene_list[0].shape[0]))

for i in range(len(ac_gene_list)):
    for idx, vector in enumerate(ac_gene_list[i]):
        ac_data_list[idx].append(torch.from_numpy(vector))                


new_ac_data_list = []
for data in ac_data_list:
    new_ac_data_list.append(torch.stack(data, dim=0).t())

```

```python
# create torch_geometric.data.Data list
def create_data_list(node_feature_mat, labels):
    # list: node_feature_mat with dim [num_drugs, -]
    # torch.Tensor for each feature with dim [num_nodes, feature_dim]
    
    data_list = []
    
    ##one-hot encoding of nodes 
#     n_values = ac_gene_list[0].shape[1]
#     values = np.arange(n_values)
#     node_onehot = np.eye(n_values)[values]
    
    if labels is not None:
        for feature, label in zip(node_feature_mat, labels):
            data_list.append(Data(x=feature.type(torch.float32), xpositional = torch.tensor(node_onehot), edge_index=edge_index, edge_attr=edge_attr.type(torch.float32), y=torch.tensor([label]).type(torch.long)))
        return data_list
    else:
        for feature in node_feature_mat:
            data_list.append(Data(x=feature.type(torch.float32),edge_index=edge_index, edge_attr=edge_attr.type(torch.float32)))
        return data_list

ac_gene_data = create_data_list(new_ac_data_list, None)

```

```python
# save drug- and food- data, and pathway
ac_data_fp = './dataset/processed/new/raw_foods.py'
import os
import os.path as osp
if not osp.exists(osp.dirname(ac_data_fp)):
    os.makedirs(osp.dirname(ac_data_fp))

torch.save(ac_gene_data, ac_data_fp)
```
